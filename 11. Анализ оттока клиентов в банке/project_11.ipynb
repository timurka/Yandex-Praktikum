# Анализ оттока клиентов: Описание и понимание задачи

**Материалы:** [Презентация (pdf)](https://drive.google.com/file/d/1OcuSZ_2wyvIT0e6kvoo_JU7bZ_Cak06_/view?usp=sharing)

**Дашборды:** [Tableau Public](https://public.tableau.com/app/profile/timur.tazetdinov/viz/_16673804310280/Dashboard1?publish=yes)

**Задача:** Проанализировать клиентов регионального банка и выделить сегменты клиентов, которые склонны уходить из банка. Почему они уходят и что их характеризует.

**Заказчик:** Менеджер продукта, который ищет способы удержать клиентов. Менеджер обладает некоторыми знаниями в статистике и математике, поэтому отчёт можно делать более детальным.

**Почему возникла задача, контекст:** менеджер продукта заметил, что уходят клиенты, но он не понимает сколько их уходят. Отдел маркетинга доложил менеджеру, что из 100 опрошенных по телефону клиентов - 70 ушли из банка.

**Зачем:** Чтобы менеджер продукта разработал меры по удержанию таких клиентов.

**Что есть:** Выгруженный датасет за весь период наблюдений. Датасет содержит данные о клиентах банка «Метанпром». Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск.

Исследования ранее проводились, но они не отвечают на этот вопрос.

**Результат исследования:** Презентация для менеджера продукта в формате pdf.

## Датасет

Поля:

- `userid` — идентификатор пользователя,
- `score` — балл кредитного скоринга, который отображает кредитоспособность и добросовестность клиента. Чем выше балл, тем надёжнее клиент.
- `City` — город,
- `Gender` — пол,
- `Age` — возраст,
- `Objects` — объекты в собственности - квартиры, машины и т.д.
- `Balance` — баланс на счёте,
- `Products` — продукты - кредитные и дебетовые карты, вклады, кредиты.
- `CreditCard` — есть ли кредитная карта,
- `Loyalty` — активный клиент - клиент, который совершает целевое действие в заданный промежуток времени (вход в приложение, транзакции, звонок на горячую линию)
- `estimated_salary` — заработная плата клиента, признак высчитываемый на основе поступлений средств клиенту.
- `Churn` — ушёл или нет.

## Распределение признаков

Посмотрим на распределения и разница в бинарных значениях.

Для начала создадим функцию отрисовки графиков:

```python
# Функция рисует гистограмму и принимает маркер оттока
def hist_churn(data_churn, title, histplot=True):
    # Располагаем графики на одном листе с колонками и рядами
    f, axes = plt.subplots(nrows=4, ncols=2, figsize=(12, 10), dpi=90, constrained_layout=True)

    # В цикле генерим графики по каждому полю
    i = 0
    j = 0

    # Перебираем поля с числовыми значениями
    for col in ['age', 'score', 'balance', 'estimated_salary', 'objects', 'products', 'city']:
        if i == 2:
            i = 0
            j += 1
            
        # Строим графики
        if histplot:
            sns.histplot(data_churn[col], ax=axes[j, i], kde=True)
        else:
            sns.boxplot(data=data, x='churn', y=col, ax=axes[j, i])
            
        i += 1
        
    f.suptitle(title, fontsize=18);
    # Уберём пустой график
    axes[3, 1].set_axis_off()
    
```

```python
hist_churn(churn1, 'Распределение признаков в группе оттока')
```

```python
hist_churn(churn0, 'Распределение признаков в группе оставшихся')
```

Разница в возрасте видна наглядно - в группе оттока центр нормального распределения находится за пределами 40 лет. В группе оставшихся более молодой возраст. При этом, в группе оттока более симметричное нормальное распределение. В группе оставшихся длинный хвост справа.

Интересно, что в группе оттока разнообразней количество продуктов на участника, чем в группе оставшихся - там четкая мода на 2 продукта. В группе оттока, в основном 1 продукт.

По распределению кредитного скоринга, видна что в группе оставшихся центр смещён правее - значение чуть больше, чем у ушедших.

В группе оставшихся больше всего из города Ярославля. В оттоке - Ярославль и маленький город в области - Ростов Великий имеют практически одинаковый процент.

```python
# Посмотрим на относительную разницу у городов

churn_city = pd.DataFrame({'city', 'per'})

i = 0
for city in data['city'].unique():
    churn_city.loc[i, 'city'] = city
    churn_city.loc[i, 'per'] = len(churn1.query('city == "'+ city +'"')) / len(data.query('city == "'+ city +'"')) * 100
    i += 1
    
print(churn_city[['city', 'per']])   
    
sns.barplot(data=churn_city, x='city', y='per', order=churn_city.sort_values('per', ascending=False)['city']).\
    set_title('Сравнение оттока в городах');
plt.xlabel("Город");
plt.ylabel("Отток, %");
```

На относительных значениях - наглядно видно, что в Ростове Великом наибольший отток (**32%** против **16%**).

Покажем сравнения более наглядно. Графики пригодятся для презентации:

```python
def compare_graph(field, title, xlabel, bins=0):
    '''
        Функция для наглядного сравнения значений
            field - поля из датафрейма для сравнения
            xlabel - подпись оси X
    '''
    if bins == 0:
        ax = churn0[field].hist(alpha=.5);
        ax = churn1[field].hist(alpha=.5);
    else:
        ax = churn0[field].hist(alpha=.5, bins=bins);
        ax = churn1[field].hist(alpha=.5, bins=bins);
    
    ax.set_title(title);
    ax.set_xlabel(xlabel);
    ax.set_ylabel('Количество');
    plt.legend(['Действующие', 'Отток'])
```

```python
compare_graph('balance', 'Сравнение баланса у групп', 'Сумма баланса, руб.')
```

```python
compare_graph('loyalty', 'Сравнение доли активных клиентов у групп', 'Активный клиент?')
```

# Проверка гипотез

```python
# Функция проверки гипотез
def test_hip(data1, data2, alpha = 0.05 ):
    results = st.mannwhitneyu(data1, data2)

    print('p-значение: ', results.pvalue)

    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу: разница статистически значима')
    else:
        print(
            'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
        ) 
```

## Средний доход

**Нулевая гипотеза**: средний доход клиентов между группами не различаются.<br />
**Альтернативная гипотеза**: средний доход клиентов между группами разный

```python
test_hip(churn0['estimated_salary'], churn1['estimated_salary'])
```

Итог: Не отвераем нулевую гипотезу - **средний доход клиентов между группами не различаются**.

С вероятностью в почти **23%** такое или большее различие можно получить случайно. Это больше установленной стат. значимости в 5%.
